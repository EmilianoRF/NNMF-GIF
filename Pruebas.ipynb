```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
%config InlineBackend.figure_format = 'svg'
fs_r = 8e3
orden = 2
fci = 80
fcs = 1200

b, a = sig.butter(orden, [fci,fcs],'bandpass',fs=fs_r)

w, h = sig.freqz(b,a,fs=fs_r)

fig,ax = plt.subplots(figsize=(15,4))
ax.plot(w, 20*np.log10(abs(h)),color='purple',linewidth=1.7)
ax.set_title('Respuesta en frecuencia (dB)')
ax.set_xlabel('Frecuencia (Hz)')
ax.set_ylabel('Magnitud (dB)')
ax.grid(True,linestyle='dashed',color='gray',alpha=0.5);
ax.set_xticks(np.arange(0,4200,200));
ax.set_xticks(np.arange(0,4100,100),minor=True);
ax.scatter(fcs,-3,color="red");
```
